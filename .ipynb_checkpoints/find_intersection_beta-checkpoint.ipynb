```python
import nibabel as nib
import numpy as np
from scipy import stats
import xml.etree.ElementTree as xml
import pandas as pd
from nilearn.plotting import plot_surf_stat_map
from matplotlib import pyplot as plt
```

```python
# https://github.com/AthenaEPI/logpar/blob/master/logpar/utils/cifti_utils.py
def extract_matrixIndicesMap(cifti_header, direction):
    ''' Retrieves the xml of a Matrix Indices Map from a cifti file.
       Parameters
       ----------
       cifti_header: cifti header
           Header of the cifti file
       direction: string
           ROW or COLUMN
       Returns
       -------
       xml_entitie
       Returns an xml (Elemtree) object
    '''
    dim = 0 if direction == 'ROW' else 1

    cxml = xml.fromstring(cifti_header.extensions[0].get_content().to_xml())

    query = ".//MatrixIndicesMap[@AppliesToMatrixDimension='{}']".format(dim)
    matrix_indices_map = cxml.find(query)

    return matrix_indices_map


def extract_brainmodel(cifti_header, structure, direction):
    '''
       Retrieves the xml of a brain model structure from a cifti file.
       Parameters
       ----------
       cifti_header: cifti header
           Header of the cifti file
       structure: string
           Name of structure
       direction: string
           ROW or COLUMN
       Returns
       -------
       xml_entitie
       Returns an xml (Elemtree) object
    '''
    matrix_indices = extract_matrixIndicesMap(cifti_header, direction)

    if structure == 'ALL':
        query = "./BrainModel"
        brain_model = matrix_indices.findall(query)
    else:
        query = "./BrainModel[@BrainStructure='{}']".format(structure)
        brain_model = matrix_indices.findall(query)
    return brain_model
```

```python
def extract_zone_convergence(sub, hemi, brain_structure, radius=0):
    # load surf
    surf_raw = nib.load('/home/fralberti/Data/HCP_zone_prim/%s/%s.%s.midthickness.32k_fs_LR.surf.gii' % (sub,sub,hemi))
    surf = []
    surf.append(surf_raw.darrays[0].data)
    surf.append(surf_raw.darrays[1].data)
    vertices, triangles = surf

    # load labels
    labels = nib.load('/home/fralberti/Data/HCP_zone_prim/%s/%s.zone_prim.32k_fs_LR.dlabel.nii' % (sub, sub))
    zones = labels.get_fdata().squeeze()

    brain_model = extract_brainmodel(labels.header, brain_structure, 'COLUMN')
    offset = int(brain_model[0].attrib['IndexOffset'])
    cortex = np.array(brain_model[0].find('VertexIndices').text.split(), dtype=int)

    z = np.zeros(vertices.shape[0])
    z[cortex] = zones[offset:offset+len(cortex)]
    # next step takes advantage of prod of 1,2,3 == 6.
    coords = np.argwhere(np.prod(z[triangles], axis=1) == 6.)

    # take more posterior node:
    trig_of_interest = np.argmin([vertices[triangles[coords[0]]][0][:,1].mean(), vertices[triangles[coords[1]]][0][:,1].mean()])
    nodes_of_interest = triangles[coords[trig_of_interest]][0]
    
    radius_tmp = 0
    while radius_tmp < radius:
        nodes_of_interest = np.append(nodes_of_interest,triangles[np.isin(triangles, nodes_of_interest).sum(1)==2])
        nodes_of_interest = np.unique(nodes_of_interest)
        radius_tmp+=1
    
    return nodes_of_interest, offset, cortex
```

```python
subject_IDs = np.load('/home/fralberti/Data/Concatenated_dtseries/subject_IDs.npy')
percentile_table = pd.DataFrame(columns=['ID', 'hemisphere', 'vertex', 'index', 'value', 'mean', 'percentile'])
# percentile_table = pd.DataFrame(columns=['ID','hemisphere',
#                                          'vertex1','vertex2','vertex3',
#                                          'index1','index2','index3',
#                                          'value1','value2','value3',
#                                          'mean','percentile'])

for sub in subject_IDs:
    hemisphere = {'L':'CIFTI_STRUCTURE_CORTEX_LEFT', 'R':'CIFTI_STRUCTURE_CORTEX_RIGHT'}
    for hemi in hemisphere.keys():
        brain_structure = hemisphere[hemi]

        nodes_of_interest, offset, cortex = extract_zone_convergence(sub, hemi, brain_structure,7)
        nodes_of_interest = np.sort(nodes_of_interest.tolist(),axis=0)

        grad1 = nib.load('/home/fralberti/Data/Gradient_1/%s_grad1.dscalar.nii' % sub)
        values = np.array(grad1.get_fdata())
        brain_model = [x for x in grad1.header.get_index_map(1).brain_models if x.brain_structure==brain_structure]
        offset = brain_model[0].index_offset
        count = brain_model[0].index_count
        vertices = np.array(brain_model[0].vertex_indices)
        idx = np.array([i for i,x in enumerate(vertices) if x in nodes_of_interest])
        node_values = values[0:,offset+idx]

        grad_percentile_of_convergence_nodes = stats.percentileofscore(values[0:,offset:offset+count].T, np.mean(values[0:,offset+idx]))

        nu_subject = pd.DataFrame(np.vstack([np.repeat(sub,len(nodes_of_interest)),
                                   np.repeat(hemi,len(nodes_of_interest)),
                                   nodes_of_interest, idx, node_values[0],
                                   np.repeat(np.mean(values[0:,offset+idx]),len(nodes_of_interest)),
                                   np.repeat(grad_percentile_of_convergence_nodes,len(nodes_of_interest))]).T,
                                  columns=percentile_table.columns)
        percentile_table = percentile_table.append(nu_subject)
#         percentile_table = percentile_table.append({'ID':sub,'hemisphere':hemi,'vertex':nodes_of_interest,
#                                                     'index':idx,'value':node_values[0],'subject_mean':np.mean(values[0:,offset+idx]),
#                                                     'subject_percentile':grad_percentile_of_convergence_nodes},ignore_index=True)
#         percentile_table = percentile_table.append({'ID':sub,'hemisphere':hemi,
#                                                     'vertex1':nodes_of_interest[0],'vertex2':nodes_of_interest[1],'vertex3':nodes_of_interest[2],
#                                                     'index1':idx[0],'index2':idx[1],'index3':idx[2],
#                                                     'value1':node_values[0][0],'value2':node_values[0][1],'value3':node_values[0][2],
#                                                     'mean':np.mean(values[0:,offset+idx]),'percentile':grad_percentile_of_convergence_nodes}, ignore_index=True)
percentile_table.reset_index(inplace=True,drop=True)
percentile_table.to_csv('/home/fralberti/Data/Gradient_1/gradient1_percentiles.csv',index=False)
del percentile_table
```

```python
subject_IDs = np.load('/home/fralberti/Data/Concatenated_dtseries/subject_IDs.npy').astype('int')
orig_percentile_df = pd.read_csv('/home/fralberti/Data/Gradient_1/gradient1_percentiles.csv')
perm_percentile_df = pd.DataFrame(columns=['ID_perm', 'ID_orig', 'hemisphere', 'vertex', 'index', 'value', 'mean', 'percentile'])
hemi_dict = {'L':'CIFTI_STRUCTURE_CORTEX_LEFT', 'R':'CIFTI_STRUCTURE_CORTEX_RIGHT'}

grad1_all = dict()
for sub in subject_IDs:
    grad1_all.update({sub:nib.load('/home/fralberti/Data/Gradient_1/%s_grad1.dscalar.nii' % sub)})

for row, sub_orig, hemi, idx in orig_percentile_df[['ID','hemisphere','index']].itertuples():
    for sub_perm in subject_IDs[subject_IDs!=sub_orig]:
        
        brain_structure = hemi_dict[hemi]
        
        grad1_dscalar = grad1_all[sub_perm]
        values = np.array(grad1_dscalar.get_fdata())
        brain_model = [x for x in grad1_dscalar.header.get_index_map(1).brain_models
                       if x.brain_structure==brain_structure]
        offset = brain_model[0].index_offset
        count = brain_model[0].index_count
        
        nodes_of_interest = np.array(brain_model[0].vertex_indices)[idx]
        node_values = values[0:,offset+idx]
        
        grad1_percentile_of_convergence_nodes = stats.percentileofscore(values[0:,offset:offset+count].T,
                                                                       np.mean(node_values))
        
        nu_subject = []
        perm_percentile_df = perm_percentile_df.append({'ID_perm':sub_perm,'ID_orig':sub_orig,'hemisphere':hemi,
                                                        'vertex':nodes_of_interest,'index':idx,'value':node_values[0]},
                                                        ignore_index=True)



perm_percentile_df = perm_percentile_df.sort_values(['ID_perm','ID_orig']).reset_index(drop=True)
perm_percentile_df['mean'] = perm_percentile_df.groupby(['ID_orig','ID_perm','hemisphere']).transform('mean')['value']
perm_percentile_df['percentile'] = [stats.percentileofscore(grad1_all[row['ID_perm']].get_fdata()[0:,offset:offset+count].T, row['mean']) 
                                    for i,row in perm_percentile_df.iterrows()]
perm_percentile_df.to_csv('/home/fralberti/Data/Gradient_1/gradient1_percentiles_perm',index=False)
perm_percentile_df
```
