```python
import nibabel as nib
import numpy as np
from scipy import stats
import xml.etree.ElementTree as xml
import pandas as pd

# https://github.com/AthenaEPI/logpar/blob/master/logpar/utils/cifti_utils.py
def extract_matrixIndicesMap(cifti_header, direction):
    ''' Retrieves the xml of a Matrix Indices Map from a cifti file.
       Parameters
       ----------
       cifti_header: cifti header
           Header of the cifti file
       direction: string
           ROW or COLUMN
       Returns
       -------
       xml_entitie
       Returns an xml (Elemtree) object
    '''
    dim = 0 if direction == 'ROW' else 1

    cxml = xml.fromstring(cifti_header.extensions[0].get_content().to_xml())

    query = ".//MatrixIndicesMap[@AppliesToMatrixDimension='{}']".format(dim)
    matrix_indices_map = cxml.find(query)

    return matrix_indices_map


def extract_brainmodel(cifti_header, structure, direction):
    '''
       Retrieves the xml of a brain model structure from a cifti file.
       Parameters
       ----------
       cifti_header: cifti header
           Header of the cifti file
       structure: string
           Name of structure
       direction: string
           ROW or COLUMN
       Returns
       -------
       xml_entitie
       Returns an xml (Elemtree) object
    '''
    matrix_indices = extract_matrixIndicesMap(cifti_header, direction)

    if structure == 'ALL':
        query = "./BrainModel"
        brain_model = matrix_indices.findall(query)
    else:
        query = "./BrainModel[@BrainStructure='{}']".format(structure)
        brain_model = matrix_indices.findall(query)
    return brain_model
```

```python
def extract_zone_convergence(sub, hemi, brain_structure):
    # load surf
    surf_raw = nib.load('/home/fralberti/Data/HCP_zone_prim/%s/%s.%s.midthickness.32k_fs_LR.surf.gii' % (sub,sub,hemi))
    surf = []
    surf.append(surf_raw.darrays[0].data)
    surf.append(surf_raw.darrays[1].data)
    vertices, triangles = surf

    # load labels
    labels = nib.load('/home/fralberti/Data/HCP_zone_prim/%s/%s.zone_prim.32k_fs_LR.dlabel.nii' % (sub, sub))
    zones = labels.get_fdata().squeeze()

    brain_model = extract_brainmodel(labels.header, brain_structure, 'COLUMN')
    offset = int(brain_model[0].attrib['IndexOffset'])
    cortex = np.array(brain_model[0].find('VertexIndices').text.split(), dtype=int)

    z = np.zeros(vertices.shape[0])
    z[cortex] = zones[offset:offset+len(cortex)]
    # next step takes advantage of prod of 1,2,3 == 6.
    coords = np.argwhere(np.prod(z[triangles], axis=1) == 6.)

    # take more posterior node:
    trig_of_interest = np.argmin([vertices[triangles[coords[0]]][0][:,1].mean(), vertices[triangles[coords[1]]][0][:,1].mean()])
    nodes_of_interest = triangles[coords[trig_of_interest]][0]
    
    return nodes_of_interest, offset, cortex
```

```python
subject_IDs = np.load('/home/fralberti/Data/Concatenated_dtseries/subject_IDs.npy')
percentile_table = pd.DataFrame(columns=['ID','hemisphere',
                                         'vertex1','vertex2','vertex3',
                                         'index1','index2','index3',
                                         'value1','value2','value3',
                                         'mean','percentile'])

for sub in subject_IDs:
    hemisphere = {'L':'CIFTI_STRUCTURE_CORTEX_LEFT', 'R':'CIFTI_STRUCTURE_CORTEX_RIGHT'}
    for hemi in hemisphere.keys():
        brain_structure = hemisphere[hemi]

        nodes_of_interest, offset, cortex = extract_zone_convergence(sub, hemi, brain_structure)
        nodes_of_interest = np.sort(nodes_of_interest.tolist(),axis=0)

        grad1 = nib.load('/home/fralberti/Data/Gradient_1/%s_grad1.dscalar.nii' % sub)
        values = np.array(grad1.get_fdata())
        b_model = [x for x in grad1.header.get_index_map(1).brain_models if x.brain_structure==brain_structure]
        offset = b_model[0].index_offset
        count = b_model[0].index_count
        vertices = np.array(b_model[0].vertex_indices)
        idx = np.array([i for i,x in enumerate(vertices) if x in nodes_of_interest])
        nodes_grad1 = values[0:,offset+idx]

        grad_percentile_of_convergence_nodes = stats.percentileofscore(values[0:,offset:offset+count].T, np.mean(values[0:,offset+idx]))

        hemi_code = (0 if hemi=='L' else 1)
#         percentile_table = percentile_table.append({'ID':sub,'hemisphere':hemi_code,'vertex1':nodes_of_interest,
#                                                     'indexes':idx,'values':nodes_grad1[0],'mean':np.mean(values[0:,offset+idx]),
#                                                     'percentile':grad_percentile_of_convergence_nodes},ignore_index=True)
        percentile_table = percentile_table.append({'ID':sub,'hemisphere':hemi_code,
                                                    'vertex1':nodes_of_interest[0],'vertex2':nodes_of_interest[1],'vertex3':nodes_of_interest[2],
                                                    'index1':idx[0],'index2':idx[1],'index3':idx[2],
                                                    'value1':nodes_grad1[0][0],'value2':nodes_grad1[0][1],'value3':nodes_grad1[0][2],
                                                    'mean':np.mean(values[0:,offset+idx]),'percentile':grad_percentile_of_convergence_nodes}, ignore_index=True)

percentile_table.to_csv('/home/fralberti/Data/Gradient_1/gradient1_percentiles',index=False)
percentile_table
#del percentile_table
```

```python
subject_IDs = np.load('/home/fralberti/Data/Concatenated_dtseries/subject_IDs.npy').astype('int')

orig_percentile_df = pd.read_csv('/home/fralberti/Data/Gradient_1/gradient1_percentiles')
perm_percentile_df = pd.DataFrame(columns=['ID_orig','ID_perm','hemisphere',
                                         'vertex1','vertex2','vertex3',
                                         'index1','index2','index3',
                                         'value1','value2','value3',
                                         'mean','percentile'])

for row, sub_orig, hemi_code, idx1, idx2, idx3 in orig_percentile_df[['ID','hemisphere','index1','index2','index3']].itertuples():
    for sub_perm in subject_IDs[subject_IDs!=sub_orig]:
        
        hemisphere = {'L':'CIFTI_STRUCTURE_CORTEX_LEFT', 'R':'CIFTI_STRUCTURE_CORTEX_RIGHT'}
        hemi = ('L' if hemi_code==0 else 'R')
        brain_structure = hemisphere[hemi]
        
        grad1 = nib.load('/home/fralberti/Data/Gradient_1/%s_grad1.dscalar.nii' % sub_perm)
        values = np.array(grad1.get_fdata())
        b_model = [x for x in grad1.header.get_index_map(1).brain_models if x.brain_structure==brain_structure]
        offset = b_model[0].index_offset
        count = b_model[0].index_count
        
        idx = np.array([idx1,idx2,idx3])
        nodes_of_interest = np.array(b_model[0].vertex_indices)[idx]
        nodes_grad1 = values[0:,offset+idx]
        
        grad_percentile_of_convergence_nodes = stats.percentileofscore(values[0:,offset:offset+count].T, np.mean(nodes_grad1))
        
        perm_percentile_df = perm_percentile_df.append({'ID_orig':sub_orig,'ID_perm':sub_perm,'hemisphere':hemi,
                                                    'vertex1':nodes_of_interest[0],'vertex2':nodes_of_interest[1],'vertex3':nodes_of_interest[2],
                                                    'index1':idx1,'index2':idx2,'index3':idx3,
                                                    'value1':nodes_grad1[0][0],'value2':nodes_grad1[0][1],'value3':nodes_grad1[0][2],
                                                    'mean':np.mean(nodes_grad1),'percentile':grad_percentile_of_convergence_nodes},
                                                   ignore_index=True)
        
perm_percentile_df.to_csv('/home/fralberti/Data/Gradient_1/gradient1_percentiles_perm',index=False)

perm_percentile_df
```
